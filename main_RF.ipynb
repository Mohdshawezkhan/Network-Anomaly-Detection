```python
Net_data_new=Net_data_new.loc[~Net_data_new.duplicated()]
```

```python
Net_data_new.shape
```

```python
Net_data_new['attack'].loc[Net_data_new['attack']!='normal']='attack'
```

```python
Obj_col=Net_data_new.select_dtypes(include='object').columns
print(Obj_col)
```

```python
le=LabelEncoder()
for i in Obj_col:
    Net_data_new[i]=le.fit_transform(Net_data_new[i])
```

```python
X=Net_data_new.drop(['attack'],axis=1)
y=Net_data_new['attack']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
rf=RandomForestClassifier(n_estimators=100)
rf.fit(X_train,y_train)
y_pred=rf.predict(X_test)
rf_score=rf.score(X_test,y_test)*100
print(rf_score)
```

```python
acc_score=accuracy_score(y_test,y_pred)
print(acc_score*100)
```
